## Import Libraries

In this step, we import required Python libraries for data analysis, visualization, and machine learning.
- pandas → data handling
- numpy → numerical operations
- matplotlib & seaborn → visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

Here, we load red wine and white wine datasets using pandas.
The separator used in dataset is ';' so we specify sep=';'.

```python
red = pd.read_csv("winequality-red.csv", sep=";")
white = pd.read_csv("winequality-white.csv", sep=";")
```

## View Data

We display first 5 rows of the dataset to understand the structure and features.

```python
print("Red Wine Data:")
print(red.head())

print("\nWhite Wine Data:")
print(white.head())
```

## Dataset Information

We check dataset information such as:
- Column names
- Data types
- Missing values

```python
print(red.info())
print(white.info())
```

## Data Visualization

We visualize the distribution of wine quality using count plots.
This helps us understand how many samples belong to each quality category.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='quality', data=red)
plt.title("Red Wine Quality Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='quality', data=white)
plt.title("White Wine Quality Distribution")
plt.show()
```

## Correlation Heatmap

Heatmap shows the relationship between different features.
Higher values indicate strong correlation.

```python
plt.figure(figsize=(10,6))
sns.heatmap(red.corr(), annot=True, cmap="coolwarm")
plt.title("Red Wine Correlation")
plt.show()
```

## Data Preparation

We separate features (X) and target variable (y).  
Target variable is **'quality'**.

```python
X_red = red.drop("quality", axis=1)
y_red = red["quality"]

X_white = white.drop("quality", axis=1)
y_white = white["quality"]
```

## Train-Test Split

We split the dataset into training and testing data.
80% data is used for training and 20% for testing.

```python
from sklearn.model_selection import train_test_split

Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_red, y_red, test_size=0.2, random_state=42)

Xw_train, Xw_test, yw_train, yw_test = train_test_split(X_white, y_white, test_size=0.2, random_state=42)
```

## Model Training

We use Random Forest Classifier to train the model.
It is an ensemble learning method that gives good accuracy.

```python
from sklearn.ensemble import RandomForestClassifier

model_red = RandomForestClassifier()
model_red.fit(Xr_train, yr_train)

model_white = RandomForestClassifier()
model_white.fit(Xw_train, yw_train)
```

## Predictions

We use trained model to predict wine quality on test data.

```python
yr_pred = model_red.predict(Xr_test)
yw_pred = model_white.predict(Xw_test)
```

## Accuracy Score

Accuracy shows how many predictions are correct out of total predictions.

```python
from sklearn.metrics import accuracy_score

print("Red Wine Accuracy:", accuracy_score(yr_test, yr_pred))
print("White Wine Accuracy:", accuracy_score(yw_test, yw_pred))
```

## Confusion Matrix

Confusion Matrix shows the performance of classification model.
It compares actual values with predicted values.

```python
from sklearn.metrics import confusion_matrix

print("Red Wine Confusion Matrix:")
print(confusion_matrix(yr_test, yr_pred))

print("\nWhite Wine Confusion Matrix:")
print(confusion_matrix(yw_test, yw_pred))
```

Confusion Matrix helps in understanding which classes are predicted correctly and where the model makes mistakes.

- Model struggles between quality 5 and 6
- White wine model performs slightly better
